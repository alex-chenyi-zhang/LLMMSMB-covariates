First load all the input files and the results of the inference algorithm

```julia
covariate_file_names = "../X_files_chr2_100k_2.txt"
map_file_names = "../Y_files_chr2_100k_2.txt"

io = open(covariate_file_names, "r")
covariate_files = readdlm(io, String)
close(io)

io = open(map_file_names, "r")
contact_map_files = readdlm(io, String)
close(io)

n_regions = length(contact_map_files)
println("n_regions = ", n_regions)

println(contact_map_files[1], "\t" , covariate_files[1])
# read the first files from list
io = open("..$(covariate_files[1][5:end])","r")
X_i = readdlm(io, Float64; header=true)[1]
close(io)

io = open("..$(contact_map_files[1][5:end])","r")
Y_i = readdlm(io, Float64)
close(io)

N = length(X_i[1,:])
P = length(X_i[:,1])

# data structures for aggregated covariates and contact maps
X = zeros(P,N*n_regions)
Y = zeros(N, N*n_regions)
X[:,1:N] .= X_i
Y[:,1:N] .= Y_i
for i_region in 2:n_regions
    println("..$(covariate_files[i_region][5:end])", "\t" , "..$(contact_map_files[i_region][5:end])")
    io = open("..$(covariate_files[i_region][5:end])","r")
    X_i = readdlm(io, Float64; header=true)[1]
    close(io)

    io = open("..$(contact_map_files[i_region][5:end])","r")
    Y_i = readdlm(io, Float64)
    close(io)
    X[:,(i_region-1)*N+1:i_region*N] .= X_i
    Y[:,(i_region-1)*N+1:i_region*N] .= Y_i
end
println(length(X[1,:]))
println(length(Y[1,:]))
println(N, "\t", P)
```

```julia
res_dir = "chr2_100k_multiregion/"
K = 9

io = open("$(res_dir)B_$(N)_$(K)_chr2_100k.txt","r")
B = readdlm(io,Float64)
close(io)

io = open("$(res_dir)Sigma_$(N)_$(K)_chr2_100k.txt","r")
Σ = readdlm(io,Float64)
close(io)

io = open("$(res_dir)Gamma_$(N)_$(K)_chr2_100k.txt","r")
Γ = readdlm(io,Float64)
close(io)

io = open("$(res_dir)like_var_$(N)_$(K)_chr2_100k.txt","r")
like_var = readdlm(io,Float64)
close(io)
```

```julia
elbows = []
lambdas = []
#mus = []


for i_region in 1:n_regions
    reg = covariate_files[i_region][27:end]
    
    io = open("$(res_dir)elbows_$(N)_$(K)$(reg)","r")
    elb_i = readdlm(io,Float64)
    close(io)
    
    io = open("$(res_dir)lambda_$(N)_$(K)$(reg)","r")
    lamb_i = readdlm(io,Float64)
    close(io)
    
    #io = open("$(res_dir)mu_$(N)_$(K)$(reg)","r")
    #mu_i = readdlm(io,Float64)
    #close(io)
    
    push!(elbows, elb_i)
    push!(lambdas, lamb_i)
    #push!(mus, mu_i)
end

n_iterations =length(elbows[1][:,1])

```

Now we can generate prediction from the MAP estimate of the topic distributions
---

```julia
# we take as topic distributions the softmax of the λ from the inference algorithm. 

thetas = []

for i_region in 1:n_regions
    theta_i = zeros(K*n_iterations, N)
    for iter in 1:n_iterations
        theta_i[K*(iter-1)+1:K*iter,:] .= softmax(lambdas[i_region][K*(iter-1)+1:K*iter,:])
    end
    push!(thetas, theta_i)
end
```

```julia
iter = 1
plot(thetas[1][K*(iter-1)+1:K*iter,:]', size = (600, 300))
```

```julia
# I will call Y_fit the map generated from the MAP thetas

Y_fit = []

for i_region in 1:n_regions
    Y_fit_i = zeros(N*n_iterations,N)
    for iter in 1:n_iterations
        for i in 1:N
            for j in 1:i-1
                z_i = sample(Weights(thetas[i_region][(iter-1)*K+1:iter*K,i]))
                z_j = sample(Weights(thetas[i_region][(iter-1)*K+1:iter*K,j]))

                rate = B[z_i+K*(iter-1),z_j] 
                Y_fit_i[i+N*(iter-1),j] = rate
                Y_fit_i[j+N*(iter-1),i] = Y_fit_i[i+N*(iter-1),j]
            end
        end
    end
    push!(Y_fit, Y_fit_i)
end
```

Last step is to generate graph instances starting directly from the biological covariates and the estimates global hyper parameters
-----

This would be the actual interesting prediction. Before we couldn't do it very well. Now that we have more regions to consider, hopefully the results will get better.

```julia
new_lambdas = []
new_thetas = []

for i_region in 1:n_regions
    new_lambda_i = zeros(K*n_iterations, N)
    new_theta_i = zeros(K*n_iterations, N)
    for iter in 1:n_iterations
        new_lambda_i[K*(iter-1)+1:K*iter,:] = Γ[K*(iter-1)+1:K*iter,:] * X[:,N*(i_region-1)+1:N*i_region]
        new_theta_i[K*(iter-1)+1:K*iter,:] .= softmax(new_lambda_i[K*(iter-1)+1:K*iter,:])
    end
    push!(new_thetas, new_theta_i)
    push!(new_lambdas, new_lambda_i)
end
```

```julia
lr = 0.01
input_size = length(X[1:end-1,1])
output_size = length(lambdas[1][1:7,1])

model = Chain(Dense(input_size, 32, tanh), Dense(32, output_size))
ps = Flux.params(model)
opt = ADAM(lr)
L(a,b) = (Flux.Losses.kldivergence(softmax(model(a)), softmax(b)))
```

```julia
function train_model!(mod, par, x, l)
    for i in 1:20
        gs = gradient(() -> L(x,l), par)
        Flux.Optimise.update!(ADAM(0.01), ps, gs)
        println(i, " loss: ", L(x,l))
    end
end
```
